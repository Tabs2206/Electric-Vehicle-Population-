#Electric Vehicle in each County
## We are visulatisation to get the electric vehicle in each county

```python
# Importing required libereary
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
dataset
```

```python
# To get top 5 county having more no. of electric vehicle
county = dataset['County'].value_counts().head(5)

# Save the modified dataset to a new CSV file
county.to_csv('county.csv', index=False)

plt.figure(figsize = (12,6))

sns.barplot(x=county.index, y=county.values)

plt.title('Top 5 Counties with Electric Vehicles')
plt.xlabel('County')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation = 45)

# Save the plot in a png file
plt.savefig('electric_vehicle_in_county.png')

#show the plot
plt.show()
```
