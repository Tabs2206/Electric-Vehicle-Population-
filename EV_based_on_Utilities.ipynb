# Distribution of electric range for top 10 utilities
## The plot illustrates
The median electric range for each of these top utilities.
The spread and variability of the ER within each utility.
Any outliers in the ER data for each utilities.

```python
#import the required liberary
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
dataset
```

```python
#identify the top 10 utilities
top10 = dataset['Electric Utility'].value_counts().head(10).index

top10
```

```python
#filtering the data for top 10 utilities
dataTop10 = dataset[dataset['Electric Utility'].isin(top10)]

dataTop10

# Save the plot in a png file
plt.savefig('datatop.png')
```

```python
#Creating the box plot for top 10 utilities

plt.figure(figsize = (15,8))
sns.boxplot(x = 'Electric Utility', y = 'Electric Range',data = dataTop10, palette = 'tab10' )
plt.title('Box Plot Electric Range for Top 10 Electric Utilities')
plt.xlabel('Electric Utilities')
plt.ylabel('Electric Range')
plt.xticks(rotation = 45)

# Save the plot in a png file
plt.savefig('boxplot.png')

plt.show()
```
