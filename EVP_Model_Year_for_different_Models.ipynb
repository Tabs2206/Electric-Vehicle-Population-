#EVP Model Year for different Models
## In this visualisation we are considering three variables which have categorical data

```python
#import the required liberary
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
dataset
```

```python
#Selecting a subset of the data for clarity
# Selecting top 5 makes and few models from each
top5 = dataset['Make'].value_counts().head(5).index
subset = dataset[dataset['Make'].isin(top5)]

# Save the modified dataset to a new CSV file
subset.to_csv('subset.csv')
subset

```

```python
#creating FacetGrid
# Each row is represting a different make
g = sns.FacetGrid(subset, col="Make",col_wrap=3, height=4, aspect=1.5)
g.map_dataframe(sns.histplot, x = "Model Year", hue = "Model", multiple = "stack", palette = "Set2", bins = 20,kde = False)
g.set_axis_labels("Model Year", "Count")
g.add_legend()
plt.subplots_adjust(top = 0.9)
g.fig.suptitle('Distribution of Model Years by Make and Model')

# Save the plot in a png file
plt.savefig('facetgrid.png')
plt.show()
```
