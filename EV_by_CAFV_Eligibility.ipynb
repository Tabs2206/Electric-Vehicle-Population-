#EV on Clean Alternative Fuel Vehcile(CAFV)
## Using data visulalisation to show the distribution or count of Vehicle based on CAFV eligibility status.

```python
#import the required liberary
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
dataset
```

```python
#creating a count plot CAFV Eligibility

plt.figure(figsize = (10,6))
sns.countplot(x = 'Clean Alternative Fuel Vehicle (CAFV) Eligibility', data = dataset, color='darkgreen')
plt.title('Count of Electric vehicles by CAFV Eligibility')
plt.xlabel('CAFV Eligibility')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation = 45)

# Save the plot in a png file
plt.savefig('countplot.png')
plt.show()
```
