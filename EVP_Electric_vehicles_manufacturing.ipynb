#Electric Vehicle Manufacturing per year

## here we try to analyse the top 5 vehicle manufacturing brands based on year data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the data i.e CSV in the dataset DF
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')

#Printing the dataset to analyze the CSV
dataset
```

```python
# Count the number of vehicles per make
make_distribution = dataset['Make'].value_counts()
#Getting the top 5 vehicles per make
make_distribution = make_distribution.head(5)
#Printing the top 5 companies which make large no. of Vehicles per make

make_distribution

# Save the modified dataset to a new CSV file
make_distribution.to_csv('make_distribution.csv', index=False)
```

```python
# Performing data visualization figure size
plt.figure(figsize=(10, 6))
# Data visualization is create on pie chart
make_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140)
# Provide the title for the
plt.title('Make Distribution of Electric Vehicles')
plt.ylabel('')  # Hide the y-label
# Save the plot in a png file
plt.savefig('electric_vehicles_by_company.png')
# Show the plot
plt.show()

```
