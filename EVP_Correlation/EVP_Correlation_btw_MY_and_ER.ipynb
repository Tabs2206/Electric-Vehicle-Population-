#Analysis of Electric Vehicles Population correlation between Model Year and Electric Range.
##Here we try to create visualization for analysis of electric vehicle population relation between model year and electric range.

```python
# Importing required libereary
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
dataset
```

```python
#Plot figure size
plt.figure(figsize=(12,6))

# Plotting a scattered grapgh
sns.scatterplot(data = dataset, x= 'Model Year', y = 'Electric Range', alpha =0.6, color = 'purple')
plt.title('Correlation Between Model Year and Electric Range')
plt.xlabel('Model Year')
plt.ylabel('Electric Range (miles)')
plt.grid(True)
plt.tight_layout()


# Save the plot in a png file
plt.savefig('electric_vehicles_types.png')

#Display the plot
plt.show()
```
