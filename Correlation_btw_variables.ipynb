#Visualisation of correlation between Numeric Variables
## here we are trying to find the correlation between the numerical variables


```python
#import the required liberary
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
dataset
```

```python
# To get the numeric variables from the dataset
numericData = dataset[['Model Year','Electric Range','Base MSRP']]
numericData

# Save the modified dataset to a new CSV file
numericData.to_csv('numeric_data.csv')
```

```python
# Creating a heatmap
correlationMatrix = dataset.corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlationMatrix,  cmap ='coolwarm')
plt.title('Heatmap of Correlation between numeric variables')

# Save the plot in a png file
plt.savefig('heat_map_correlation.png')

plt.show()
```
