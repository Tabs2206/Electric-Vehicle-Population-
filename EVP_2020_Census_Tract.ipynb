#EVP 2020cCensus Tracts
## The distribution of electric vehicle counts across different 2020 Census Tracts. This would give us an idea of how densely electric vehicles are distributed in different tracts.

```python
#import the required liberary
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
dataset
```

```python
# Counting the number of electric vehicles per census tract
count = dataset['2020 Census Tract'].value_counts()
count


# Save the modified dataset to a new CSV file
count.to_csv('count.csv')
```

```python
# Density Plot
plt.subplot(1, 2, 2)
sns.kdeplot(count, shade=True, color='green')
plt.title('Density Plot of EV Counts per 2020 Census Tract')
plt.xlabel('Number of Electric Vehicles')
plt.ylabel('Density')

plt.tight_layout()


# Save the plot in a png file
plt.savefig('densityplot.png')

plt.show()
```
