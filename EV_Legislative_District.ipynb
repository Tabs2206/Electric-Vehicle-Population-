#EV Legislative District
##  The trend of the average electric range of electric vehicles across different model years.

```python
#import the required liberary
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
dataset
```

```python
# Calculating the average electric range per model year
averageRange = dataset.groupby('Model Year')['Electric Range'].mean()


# Save the modified dataset to a new CSV file
averageRange.to_csv('average_range.csv')

averageRange
```

```python
# Plotting linear plot
plt.figure(figsize=(12, 6))
averageRange.plot(kind='line', color='purple', marker='o')
plt.title('Trend of Average Electric Range Over Model Years')
plt.xlabel('Model Year')
plt.ylabel('Average Electric Range')
plt.grid(True)
plt.tight_layout()


# Save the plot in a png file
plt.savefig('avarage_range.png')

plt.show()
```
